```python
import pandas as pd
```

## Location Identifiers

Read derived ONS/OS location data

```python
#read Derived Location Data
dfLoc = pd.read_csv('Derived Data/Location_Data.csv') 
```

## EPC Certificates

Develop Housing database from https://epc.opendatacommunities.org/

```python
#read EPC raw data
dfEPC = pd.read_csv('Source Data/Housing/EPC certificates.csv') 
```

```python
#add location identifiers (merge on Unique property reference number)
dfEPC_Loc = pd.merge(dfEPC,dfLoc, on='UPRN', how='left')
```

# Add Gas and electricity

Add postcode level gas and electricity from

https://www.gov.uk/government/collections/sub-national-electricity-consumption-data

https://www.gov.uk/government/collections/sub-national-gas-consumption-data

```python
#read data
dfGas21 = pd.read_csv('Source Data/Housing/Postcode_level_gas_2021.csv') 
```

```python
dfElec21 = pd.read_csv('Source Data/Housing/Postcode_level_standard_electricity_2021_K_to_Z.csv') 
```

```python
#merge utlities data
dfUtilities21 = pd.merge(dfElec21,dfGas21, on ='Postcode', how='left')
```

```python
#add year
dfUtilities21['YEAR']=2021
```

```python
#combine gas data with EPC dataset
dfEPCfull = pd.merge(dfEPC_Loc, dfUtilities21, left_on=['POSTCODE'], right_on=['Postcode'], how='left')
```

```python
#reset index to UPRN
dfEPCfull.set_index('UPRN')
```

```python
#save to excel
dfEPCfull.to_csv('Derived Data/Combined_Datasets.csv')
```
