## Merge Dataframes

```python
dflist = [pMC, pWC, pLC,pURC,pPTy,pPTe,pPA,pNoS,pHR,pWCon,pWIPn,pEZ,pCA,pFR,pFC,pFI,pLIP,pRIR,pGT,pMFT,pMHS,pSFT,pHC,pSAP,pMT,pCO2B,
         pSPI,pSTI,pASHP,pGSHP,pBBI,pWTI,pOGG,pCWI,pSWI,pLIF,pFIF,pGF,pHPF,pSF,pMEAS,pIMDO,pIMDD,pIMDI,pIMDR,pIMDHe,pIMDEd,pIMDEn,
          pIMDHo,pCPH,pCLM,pMSI,pMSIb,pFPF,pEAWE,pEFFPQ,pLADFE,
          FAE,FP,AHMin,AHMax,AH,RHMax,RH,
         SAPE,EDE,FBE,CO2,PVP,WS,DGG,CBCl,LFPP]
```

```python
final_df = reduce(lambda  left,right: pd.merge(left,right,on=['COA_CODE'],
                                            how='outer'), dflist)
```
